# 1.Import libraries & load clean data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv( r"C:\Users\viren\OneDrive\Desktop\Road accident data analytics\archive (6)\clean_accidents.csv")

df.head()
```

# 2.Accidents by Severity (Bar Chart)

```python
plt.figure(figsize=(6,4))
df['Accident Severity'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Accidents by Severity")
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.show()
```

# 3. Accidents by Time of Day

```python
plt.figure(figsize=(6,4))
df['Time of Day'].value_counts().plot(kind='bar', color='orange')
plt.title("Accidents by Time of Day")
plt.xlabel("Time Period")
plt.ylabel("Number of Accidents")
plt.show()
```

# 4. Accidents by Weather Conditions

```python
plt.figure(figsize=(6,4))
df['Weather Conditions'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Accidents by Weather Conditions")
plt.xlabel("Weather")
plt.ylabel("Number of Accidents")
plt.show()
```

# 5.Accidents by Vehicle Type

```python
plt.figure(figsize=(7,4))
df['Vehicle Type Involved'].value_counts().plot(kind='bar', color='purple')
plt.title("Accidents by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Accidents")
plt.show()
```

# 6. Accidents by Year

```python
plt.figure(figsize=(6,4))
df['Year'].value_counts().sort_index().plot(kind='line', marker='o', color='red')
plt.title("Accidents by Year")
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.grid(True)
plt.show()
```

# 7. Accidents by Month

```python
plt.figure(figsize=(6,4))
df['Month'].value_counts().sort_index().plot(kind='line', marker='o', color='blue')
plt.title("Accidents by Month")
plt.xlabel("Month (1â€“12)")
plt.ylabel("Number of Accidents")
plt.grid(True)
plt.show()
```
