# 1. Import Libraries and Load the Dataset

```python
import pandas as pd

# Load the raw dataset
df = pd.read_csv( r"C:\Users\viren\OneDrive\Desktop\Road accident data analytics\archive (6)\accident_prediction_india.csv")

df.head()
```

# 2. Check Column Names and Missing Values

```python
# See all columns
print("Columns in dataset:")
print(df.columns)

# Check missing values
print("\nMissing Values:")
df.isnull().sum()
```

# 3. Clean Text Columns (Remove Spaces + Standardize Case)

```python
# Columns that contain categorical text
columns_to_clean = [
    'Accident Severity',
    'Vehicle Type Involved',
    'Weather Conditions',
    'Road Type',
    'Road Condition',
    'Lighting Conditions'
]

# Clean each column
for col in columns_to_clean:
    df[col] = df[col].astype(str).str.strip().str.title()
```

 # 4. Clean 'Time of Day' Column

```python
df['Time of Day'] = df['Time of Day'].astype(str).str.strip().str.title()
```

# 5. Clean Driver-Related Columns

```python
df['Driver Gender'] = df['Driver Gender'].astype(str).str.strip().str.title()
df['Alcohol Involvement'] = df['Alcohol Involvement'].astype(str).str.strip().str.title()
df['Driver License Status'] = df['Driver License Status'].astype(str).str.strip().str.title()
```

# 6. Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

# 7. Convert Month Name → Month Number

```python
df['Month'] = pd.to_datetime(df['Month'], format='%B').dt.month
```

# 8. Convert Year to Integer

```python
df['Year'] = df['Year'].astype(int)
```
